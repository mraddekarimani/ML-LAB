```python
# 1. Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB, BernoulliNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
df=pd.read_csv('nlp_train.csv')
print(df["text"].iloc[4])
```

```python
# 2. Import 20 Newsgroups dataset from scikit-learn datasets
categories = ['alt.atheism','comp.graphics','sci.space']
```

```python
print("5",)
```

```python
# 3. Load 20 Newsgroups train subset
newsgroups_train = fetch_20newsgroups(subset='train', categories=categories, remove=('headers', 'footers', 'quotes'))
```

```python
# 4. Load 20 Newsgroups test subset
newsgroups_test = fetch_20newsgroups(subset='test', categories=categories, remove=('headers', 'footers', 'quotes'))
```

```python
# 5. Print all target labels
print("Target labels:", newsgroups_train.target_names)
```

```python
# 6. Prepare subset of categories
print("Subset categories:", categories)
```

```python
# 7. Load train subset with three categories
train_data = newsgroups_train.data
train_labels = newsgroups_train.target
```

```python
# 8. Load test subset with three categories
test_data = newsgroups_test.data
test_labels = newsgroups_test.target
```

```python
# 9. Print new training set target names
print("Training Set Target Names:", newsgroups_train.target_names)
```

```python
# 10. Print news training data of the 5th article
print("\n5th Training Data Sample:\n", train_data[4])
```

```python
# 11. Print shape of data and targets
print("\nTrain Data Shape:", len(train_data))
print("Train Target Shape:", len(train_labels))
print("Test Data Shape:", len(test_data))
print("Test Target Shape:", len(test_labels))
```

```python
# 12. Print training set filenames (not available in fetch_20newsgroups)
# Skipping as filenames are not included in the dataset
```

```python
# 13. Convert train data into numerical format using CountVectorizer
vectorizer = CountVectorizer(stop_words='english')
X_train_counts = vectorizer.fit_transform(train_data)
```

```python
# 14. Use BernoulliNB for training
bnb = BernoulliNB()
bnb.fit(X_train_counts, train_labels)
```

```python
# 15. Convert test data into numerical format
X_test_counts = vectorizer.transform(test_data)
```

```python
# 16. Predict target labels for the test set
predictions = bnb.predict(X_test_counts)
```

```python
# 17. Find accuracy score on the test set
accuracy = accuracy_score(test_labels, predictions)
print("\nBernoulliNB Accuracy:", accuracy)
```

```python
# 18. Use TfidfVectorizer instead of CountVectorizer and use MultinomialNB
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
X_train_tfidf = tfidf_vectorizer.fit_transform(train_data)
X_test_tfidf = tfidf_vectorizer.transform(test_data)

mnb = MultinomialNB()
mnb.fit(X_train_tfidf, train_labels)
```

```python
# 19. Find test set accuracy for MultinomialNB
mnb_predictions = mnb.predict(X_test_tfidf)
mnb_accuracy = accuracy_score(test_labels, mnb_predictions)
print("\nMultinomialNB Accuracy:", mnb_accuracy)
```

```python
# 20. Try avoiding stopwords and repeat the same
vectorizer_no_stopwords = CountVectorizer()
X_train_no_stopwords = vectorizer_no_stopwords.fit_transform(train_data)
X_test_no_stopwords = vectorizer_no_stopwords.transform(test_data)

bnb_no_stopwords = BernoulliNB()
bnb_no_stopwords.fit(X_train_no_stopwords, train_labels)
predictions_no_stopwords = bnb_no_stopwords.predict(X_test_no_stopwords)

accuracy_no_stopwords = accuracy_score(test_labels, predictions_no_stopwords)
print("\nBernoulliNB Accuracy without stopwords:", accuracy_no_stopwords)
```
